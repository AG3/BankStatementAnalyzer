```python
import pyecharts

def LoadAccountStatement(filename):
    f = open(filename,'r')
    accounts = {}
    cur_acc_name = ''
    for i in f:
        if i.find('^') == -1:
            cur_acc_name = i[:len(i)-1]
            accounts[cur_acc_name] = {'date':[],'description':[],'amount':[],'balance':[]}
            continue
        p = i.split('^')
        if p[2][0]!='-':
            continue
        accounts[cur_acc_name]['date'].append(p[0])
        accounts[cur_acc_name]['description'].append(p[1])
        accounts[cur_acc_name]['amount'].append(float(p[2].replace(',','')))
        accounts[cur_acc_name]['balance'].append(float(p[3].replace(',','')))
    f.close()
    return accounts


def GetBalanceByDay(account):
    date = account['date']
    cost = account['balance']
    res_x = []
    res_y = []
    curDate = ''
    for i in range(len(date)):
        if date[i] != curDate and i != 0:
            res_x.append(date[i])
            curDate = date[i]
            res_y.append(cost[i])
    return res_x,res_y

def LoadClaasificationKeyWords(filename,level=1):
    import re
    f = open(filename)
    keywords_pat = re.compile('(?<=\").*?(?=\")')
    for line in f:
        curName = []
        lv = 0
        for j in range(len(line)):
            if line[j]!='-':
                break
            lv+=1
        d = line[j:]
        s = d.split(':')
        if len(s) == 1:
            curName.append(d)
            continue
        
        
    return dic

AssortDictionray = LoadClaasificationKeyWords('newConfig')

def CalculateAssortedSum(cat_dic, accountName):
    l = len(accounts[accountName]['date'])

    cat_sum=dict()
    for i in cat_dic:
        cat_sum[i['name']] = 0
    for i in range(l):
        cated = False
        for j in cat_dic:
            flg = False
            for k in j['keys']:
                if accounts[accountName]['description'][i].lower().find(k) != -1:
                    cat_sum[j['name']] += accounts[accountName]['amount'][i]
                    flg = True
                    cated = True
                    break
        if flg is True:
            break
        if cated is False:
            cat_sum['other'] += accounts[accountName]['amount'][i]
    return cat_sum


def PlotBalanceLineChart(x,y,img=False,filename='',folder=''):
    folder += 'line/'
    lineChart = pyecharts.Line()
    lineChart.add('Balance',x_axis=x,y_axis=y)
    if img is True:
        lineChart.render(path=folder+filename,pixel_ratio=1.5)

def PlotAssortedPieChart(data,img=False,filename='',folder=''):
    folder += 'pie/'
    pie = pyecharts.Pie()
    attr = []
    v = []
    for i in data:
        attr.append(i)
        v.append(round(cat_sum[i]*-1,2))
    pie.add('',attr,v,is_legend_show=False, is_label_show=True)
    if img is True:
        pie.render(path=folder+filename,pixel_ratio=1.5)
```
